## Selenium

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

url = 'https://www.falabella.com.co/falabella-co'
driver = webdriver.Firefox()
driver.get(url)
time.sleep(6)

results = driver.find_elements(By.XPATH,"//div[@class='d-flex Card-module_card__AIGXd Card-module_card-count-4__96RUe Card-module_is-rounded__c6FV6 dy-class-collapsecard']/a")
urls = [a.get_attribute('href') for a in results]
driver.close()
```

```python
categorias = []
marcas = []
descripciones = []
precios = []
links = []

for url in urls:
    driver = webdriver.Firefox()
    driver.get(url)
    time.sleep(6)
    results = driver.find_elements(By.CLASS_NAME,'jsx-1833870204.jsx-3831830274.pod-details.pod-details-4_GRID.has-stickers')
    productos = [producto.text for producto in results]
    marcas.extend([marca.split("\n")[0] for marca in productos])
    descripciones.extend([descripcion.split("\n")[1] for descripcion in productos])
    results = driver.find_elements(By.CLASS_NAME,'jsx-1833870204.jsx-3831830274.pod-summary.pod-link.pod-summary-4_GRID')
    valores = [valor.text for valor in results]
    precios.extend([precio.split("\n")[0] for precio in valores])
    links.extend([a.get_attribute('href') for a in results])
    titulo = driver.find_element(By.TAG_NAME,'h1')
    categorias.extend([titulo.text]*len(valores))
    driver.close()
```

```python
import pandas as pd

d = {'Categoria': categorias, 'Marca': marcas, 'Producto': descripciones, 'Precio': precios, 'URL': links}

df = pd.DataFrame(data=d)

df.head(50)
```

```python
import webbrowser

for i in range(5):
    webbrowser.open_new_tab(df['URL'][i])
```
